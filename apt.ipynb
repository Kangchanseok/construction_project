```python
import pandas as pd
import numpy as np
%precision %.3f
```

```python
df = pd.read_excel(r'..\..\..\apt.xlsx',skiprows =[0])
df.head()
```

```python
# (공백이랑 (주) 제거 )
df['시공사'] = df['시공사'].str.strip()
df['시공사'] = df['시공사'].str.replace('(주)','')
df_sigongsa = df['시공사']
df_cnt = df_sigongsa.value_counts()
top50 = df_cnt[25:50]


# 시공횟수 많은 top 50 기업 
pd.DataFrame(top50)
```

```python
df2 = pd.read_excel(r'..\..\..\point3.xlsx')
df2 = df2.drop(df2.columns[[0,2,3,4,5,7]], axis=1) 
df2 = df2.rename(columns = {'Unnamed: 6' : '현장벌점 합'})
df2['업체명'] = df2['업체명'].str.strip()
df2['업체명'] = df2['업체명'].str.replace('(주)','')

# 이름 겹치는 기업들 다 합쳐주기
df2 = df2.groupby('업체명').sum()


# top50 기업중 벌점 존재하는 기업들 (이름, 시공횟수, 벌점)
cnt = 0
for i in top50.index:
    if i in df2.index:
        point = df2[df2.index == i].values
        print(i,top50[i],point[0][0])
        cnt += 1
print(cnt)
```
